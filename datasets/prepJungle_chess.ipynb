# jungle_chess_2pcs_raw_endgame_complete convertion

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/18631418/jungle_chess_2pcs_raw_endgame_complete.arff')
```

```python
data
```

```python
c='w'
target = 'class'
target_vals = ['w','b','d']

class_ = data[data[target]==c]
class_[target] = 1

other = data[data[target]!=c]
other[target] = 0
```

```python
other
```

```python
class_
```

```python
class_.drop(class_.index[:1305], inplace = True)
```

```python
class_
```

```python
new = pd.concat([class_, other], ignore_index=True)
new = new.sample(frac=1).reset_index(drop=True)
new
```

```python
new.to_csv('jungle_chess_2pcs_raw_endgame_complete-binary.csv', index=False)
```
