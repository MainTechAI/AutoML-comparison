# kc1 convertion

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/53950/kc1.arff')
```

```python
data
```

```python
def foo(row):
    if(row==True):
        return 1
    else:
        return 0

data['defects']=data.defects.apply(foo)
data
```

```python
data.to_csv('kc1.csv', index=False)
```
