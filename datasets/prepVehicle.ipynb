# Vehicle preparation

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/54/dataset_54_vehicle.arff')
```

```python
data
```

```python
c='saab'
target = 'Class'
target_vals = ['opel','saab','bus','van']

class_ = data[data[target]==c]
class_[target] = 'saab'

other = data[data[target]!=c]
```

```python
other
```

```python
class_
```

```python
cat = 0

for cat in target_vals:
    if(cat!=c):
        index = other[other[target] == cat].index[:137]
        other.drop(index, inplace = True)    


other[target] = 'other'

other.drop(other.index[:1], inplace = True)
```

```python
other
```

```python
new = pd.concat([class_, other], ignore_index=True)
new = new.sample(frac=1).reset_index(drop=True)
new
```

```python
new.to_csv('vehicle-binary.csv', index=False)
```
