# connect-4 preparation (balanced number of rows in class)

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/4965243/connect-4.arff')
```

```python
data
```

```python
c=2
target = 'class'
target_vals = [0,1,2]

class_ = data[data[target]==c]
class_[target] = 1

other = data[data[target]!=c]
other[target] = 0
```

```python
other
```

```python
class_
```

```python
class_.drop(class_.index[:21389], inplace = True)
```

```python
class_
```

```python
new = pd.concat([class_, other], ignore_index=True)
new = new.sample(frac=1).reset_index(drop=True)
new
```

```python
new.to_csv('connect-4-balanced-binary.csv', index=False)
```
