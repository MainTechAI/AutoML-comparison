```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
path = './multy-class/cifar-10.csv'
```

```python
data = pd.read_csv(path)
```

```python
# 60000 x 3072
```

```python
c=9
class_ = data[data['class']==c]
class_['class'] = 1
other = data[data['class']!=c]
```

```python
class_
```

```python
cat = 0

for i in range(10):
    if(cat!=c and cat<10):
        index = other[other['class'] == cat].index[:5333]
        other.drop(index, inplace = True)    
    cat+=1

other['class'] = 0

other.drop(other.index[:3], inplace = True)   
```

```python
other
```

```python
new = pd.concat([class_, other], ignore_index=True)
new = new.sample(frac=1).reset_index(drop=True)
new
```

```python
new.to_csv('cifar-10-binary.csv', index=False)
```
